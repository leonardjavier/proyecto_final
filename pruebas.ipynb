# analisis de datos

```python
# librerias
import pandas as pd
```

```python
# carga el data frame
df = pd.read_csv("df_restaurantes")
```

```python
# presentacion de los datos 
df
```

```python
# informacion del data frame
print(df.info())
```

```python
# descripcion del data frame
print(df.describe())
```

```python
# crecimiento de los restaurates 
df.plot.scatter(y="height",x="birth_year")
```

```python
# nivel de bebida
bebida = df.drink_level.value_counts()
bebida = bebida.sort_values(ascending=True)
bebida.plot.bar()
```

```python
# Restaurantes de los cuales te permiten fumar
fumadores_en_restaurantes = df.smoker.value_counts()
fumadores_en_restaurantes.plot.pie(autopct='%1.1f%%')
```

```python
# colores de los restaurantes 
color_del_restaurantes = df.color.value_counts()
color_del_restaurantes = color_del_restaurantes.sort_values(ascending=True)
color_del_restaurantes.plot.bar()
```

```python
# actividades 
actividades = df.activity.value_counts()
actividades.plot.pie(autopct='%1.1f%%')
```
